```python
# 1. Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Display settings

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

```python
# 3. Load Dataset

df = pd.read_csv("/content/CloudWatch_Traffic_Web_Attack.csv")
```

```python
# 4. Basic Dataset Information

print("Shape of dataset:", df.shape)
```

```python
# 5. Statistical Summary

df.describe(include='all')
```

```python
# 6. Check Missing Values

missing_values = df.isnull().sum()
print(missing_values)
```

```python
# 7. Duplicate Records

print("Duplicate rows:", df.duplicated().sum())
```

```python
# 8. Data Types & Unique values

df.nunique().sort_values(ascending=False)
```

```python
# =============================
# Exploratory Data Analysis (EDA)
# =============================
```

```python
# 9. Univariate Analysis

num_cols = df.select_dtypes(include=np.number).columns

df[num_cols].hist(figsize=(15, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts().head())
```

```python
# 10. Bivariate Analysis

# ---- 10.1 Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# ---- 10.2 Boxplot (Outlier Detection)
for col in num_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# 11. Country-based Interaction Analysis

plt.figure(figsize=(15, 8))
sns.countplot(y='src_ip_country_code', data=df,
order=df['src_ip_country_code'].value_counts().index)
plt.title('Interaction Count by Source IP Country Code')
plt.show()
```

```python
# 12. Suspicious Activities Based on Ports

plt.figure(figsize=(12, 6))
sns.countplot(x='dst_port', data=df[df['detection_types'] ==
'Suspicious'], palette='coolwarm')
plt.title('Suspicious Activities Based on Destination Port')
plt.xticks(rotation=45)
plt.show()
```
